## Алгоритм Евкліда

```python
# finds the GCD (greatest common divisor) of two positive integers
def gcd(a,b):
    if a == 0:
        return b
        
    while b != 0:
        if a > b:
            a = a - b
        else:
            b = b - a
        
    return a


# checks if given string str represents positive integer
def is_positive_integer(str):
    return str.isdecimal() and int(str) > 0

def main():
    error = False
    
    a_str = input("Enter a (positive integer): ")
    b_str = input("Enter b (positive integer): ")
    
    if is_positive_integer(a_str) == False :
        print("ERROR: a must be a positive integer")
        error = True
    
    if is_positive_integer(b_str) == False :
        print("ERROR: b must be a positive integer")
        error = True

    if error == False :
        a = int(a_str)
        b = int(b_str)
        print("GCD(a,b) = " + str(gcd(a,b)))
```
